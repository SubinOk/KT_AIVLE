```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
x = cancer.data
y = cancer.target
```

```python
x.shape, y.shape
```

```python
# train / test set 분리
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=2022)
```

```python
#####################
# 라이브러리 불러오기
#####################
import tensorflow as tf
from tensorflow import keras
```

```python
########
# 모델링
########
keras.backend.clear_session()

il = keras.layers.Input(shape=(30,))
ol = keras.layers.Dense(1, activation='sigmoid')(il)

model = keras.models.Model(il, ol)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')

model.summary()
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs=10)
```

```python
###########
# 모델 예측
###########
import numpy as np

pred_y = model.predict(x_test).reshape(-1)
pred_y = np.where(pred_y >= 0.5, 1, 0)
print(pred_y[:10])
print(y_test[:10])
```

# 히든 레이어 추가하여 모델링!

    히든 레이어 2개를 추가하여 다시 모델링!
    히든 레이어 노드 수 : 32
    히든 레이어 activation = 'relu'

```python
########
# 모델링
########
keras.backend.clear_session()

il = keras.layers.Input(shape=(30,))
d1 = keras.layers.Dense(32, activation = 'relu')(il)
d2 = keras.layers.Dense(32, activation = 'relu')(d1)
ol = keras.layers.Dense(1, activation='sigmoid')(d2)

model = keras.models.Model(il, ol)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')

model.summary()
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs=10)
```

```python
###########
# 모델 예측
###########
pred_y = model.predict(x_test).reshape(-1)
pred_y = np.where(pred_y >= 0.5, 1, 0)
print(pred_y[:10])
print(y_test[:10])
```
