# 멀티클래스 분류 연습 (over 2진분류)

![iris_flower](https://www.embedded-robotics.com/wp-content/uploads/2022/01/Iris-Dataset-Classification-1024x367.png)

```
from sklearn.datasets import load_iris
```

```
iris = load_iris()
```

```
x = iris.data # sepal, petal -> length, width (꽃잎과 꽃받침의 너비와 길이)
y = iris.target
```

```
x.shape, y.shape
```

```
y # integer Encoding
```

```
iris.target_names
```

```
# One-Hot Encoding
from tensorflow.keras.utils import to_categorical
```

- to_categorical()은 반복실행할 경우 계속해서 차원 수를 늘려감

```
y = to_categorical(y, 3) # class 개수 명시하지 않으면 자동으로 분류
```

```
x.shape, y.shape
```

# 1번. Sequential API style로 모델링 하세요

    loss = 'categorical_crossentropy' 사용!

```
import tensorflow as tf
from tensorflow import keras
```

```
keras.backend.clear_session()

model = keras.models.Sequential()
model.add(keras.layers.Dense(3, input_shape=(4,), activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics='accuracy')
```

```
model.fit(x, y, epochs=10)
```

```
y_pred = model.predict(x)
print(y_pred[:10]) # round()로 값을 처리할 경우 각 클래스의 예측 확률이 비슷하면 전부 0으로 나올 수 있음
```

```
y_pred[:10].argmax(axis=1) # 각 행에서 가장 값이 큰 열
```

# 2번. Functional API style로 모델링 하세요

```
x.shape, y.shape
```

```
keras.backend.clear_session()

il = keras.layers.Input(shape=(4,))
ol = keras.layers.Dense(3, activation='softmax')(il)

model = keras.models.Model(il, ol)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics='accuracy')
```

```
model.fit(x, y, epochs=10)
```

```
print(model.predict(x[:10]))
```
