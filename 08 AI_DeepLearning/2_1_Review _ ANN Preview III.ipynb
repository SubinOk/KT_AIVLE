```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
x.shape, y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2, random_state=2022)
```

```python
#####################
# 라이브러리 불러오기
#####################
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

- 전처리 시 train-test 분리 후 one-hot encoding 진행

```python
#######################
# 어떤 전처리 작업 필요
# Hint : y.shape
#######################
class_n = len(np.unique(y_train))

y_train = keras.utils.to_categorical(y_train, class_n)
```

```python
########
# 모델링
########
keras.backend.clear_session()

model = keras.models.Sequential()
model.add(keras.layers.Dense(3, input_shape=(4,), activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics='accuracy')

model.summary()
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs=10)
```

```python
###########
# 모델 예측
###########
y_pred = model.predict(x_train)
print(y_pred[:10].argmax(axis=1))
print(y_test[:10])
```

# 히든 레이어 추가하여 모델링! (Functional API)

    히든 레이어 2개를 추가하여 다시 모델링!
    히든 레이어 노드 수 : 32
    히든 레이어 activation = 'relu'

```python
########
# 모델링
########
keras.backend.clear_session()

il = keras.layers.Input(shape=(4,), name='Input')
h1 = keras.layers.Dense(32, activation='relu', name='Hidden_1')(il)
h2 = keras.layers.Dense(32, activation='relu', name='Hidden_2')(h1)
ol = keras.layers.Dense(3, activation='softmax', name='Output')(h2)

model = keras.models.Model(il, ol)

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics='accuracy')

model.summary()
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs=10)
```

```python
###########
# 모델 예측
###########
y_pred = model.predict(x_train)
print(y_pred[:10].argmax(axis=1))
print(y_test[:10])
```
