- 전처리

```python
# train test 분리
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# min max scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

- 모델링

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.callbacks import EarlyStopping
```

```python
keras.backend.clear_session()

model = keras.models.Sequential()
model.add(keras.layers.Dense(64, input_shape=(8, ), activation='relu'))
model.add(keras.layers.Dense(64, activation='relu'))
model.add(keras.layers.Dense(1))

model.compile(loss='mse', optimizer='adam')

model.summary()
```

```python
es = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=1, restore_best_weights=True)
```

```python
hist = model.fit(x_train, y_train, validation_split=.2, callbacks=[es], verbose=1, epochs=50)
```

```python
model.evaluate(x_test, y_test)
```
