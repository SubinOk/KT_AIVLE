```python
wine = load_wine()
```

```python
print(wine.DESCR)
```

```python
x = wine.data
y = wine.target
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import EarlyStopping
```

```python
from tensorflow.keras.utils import to_categorical
import numpy as np

n_class = len(np.unique(y_train))
y_train = to_categorical(y_train, n_class)
y_test = to_categorical(y_test, n_class)
```

```python
keras.backend.clear_session()

il = Input(shape=(13, ))
ol = Dense(3, activation='softmax')(il)

m1 = keras.models.Model(il, ol)
m1. compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

m1.summary()
```

```python
keras.backend.clear_session()

inputs = Input(shape=(13, ))
x = inputs

x = Dense(64, activation='relu')(x)
x = Dense(64, activation='relu')(x)
x = Dense(128, activation='relu')(x)

outputs = Dense(3, activation='softmax')(x)

m2 = keras.models.Model(inputs, outputs)
m2. compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

m2.summary()
```

```python
es = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=1, restore_best_weights=True)
```

```python
m1.fit(x_train, y_train, validation_split=.1, callbacks=[es], verbose=1, epochs=50)
```

```python
m2.fit(x_train, y_train, validation_split=.1, callbacks=[es], verbose=1, epochs=50)
```
